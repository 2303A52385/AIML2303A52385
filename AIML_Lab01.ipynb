#Part 2 – Visualization of Data using matplotlib, pyplots Packages[CO2]


```python
people = ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40, 45]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of patients")
plt.xlabel("Weight of patients in Kgs")
plt.ylabel("Height in Meters")
plt.show()
```

## Generate Bar PLot

```python
plt.figure(figsize = (12,4))
plt.title("people's weight in Kgs", fontsize = 16, fontstyle='italic')

plt.bar(x=people, height=weight, width=0.2,
        color='blue', edgecolor='black', alpha=0.6)

plt.xlabel("people", fontsize=15)
plt.xticks(fontsize=14, rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight (in kgs)", fontsize=15)
plt.show()
```

## Generate Histogram

```python
import numpy as np
plt.figure(figsize=(7,5))

# Main plot function 'hist'
plt.hist(weight,color='red',edgecolor='k', alpha=0.75,bins=5)

plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
